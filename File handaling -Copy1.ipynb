```python
import os
print(os.getcwd())
```

```python
#open file
f=open("demofile.txt")
print(f.read())

#closing file 
f.close()
```

```python
# with statement
with open("demofile.txt") as f:
    print(f.read())

#closing file 
f.close()
```

```python
#reading part of files
with open("demofile.txt") as f:
    print(f.read(19))
```

```python
#reading line
with open("demofile.txt") as f:
    print(f.readline())
```

```python
# calling readline two times 
with open("demofile.txt") as f:
    print(f.readline())
    print(f.readline())
```

```python
# read line by loop
with open("demofile.txt") as f:
    for x in f:
        print(x)
```

```python
# write an existing file
with open("demofile.txt", "a") as f:
  f.write("Now the file has more content!")

#open and read the file after the appending:
with open("demofile.txt") as f:
  print(f.read())
```

```python
# over writing existing file 
with open("demofile.txt", "w") as f:
  f.write("Woops! I have deleted the content!")

#open and read the file after the overwriting:
with open("demofile.txt") as f:
  print(f.read())
```

```python
# delete file 
import os
os.remove("demofile.txt")
```

```python
# checked if file exist
import os
if os.path.exists("demofile.txt"):
  os.remove("demofile.txt")
else:
  print("The file does not exist")
```
